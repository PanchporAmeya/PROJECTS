```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Amounts in millions, rev_m is revenue mean

rev_m = 170
rev_stddev = 20
iterations = 1000
```

```python
rev = np.random.normal (rev_m,rev_stddev,iterations)
rev
```

```python
plt.figure(figsize = (15,6))
plt.plot(rev)
plt.xlabel("Iterations")
plt.ylabel("Revenue Deviations")
plt.show()
```

```python
COGS = - (rev * np.random.normal(0.6,0.1))
COGS
```

```python
plt.figure(figsize = (15,6))
plt.plot(COGS)
plt.xlabel("Iterations")
plt.ylabel("COGS Deviations")
plt.show()
```

```python
COGS.mean()
```

```python
COGS.std()
```

```python
Gross_Profit = rev + COGS
Gross_Profit
```

```python
plt.figure(figsize = (15,6))
plt.plot(Gross_Profit)
plt.xlabel("Iterations")
plt.ylabel("Gross Profit Deviations")
plt.show()
```
