```python
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
tickers = ['JPM','BAC','MS','C','WFC','^VIX','^TNX','DX-Y.NYB','CL=F','GC=F','SPY']
data = yf.download(tickers, start = '2002-01-01', end ='2025-01-01')['Close']
data
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (16,9))
plt.plot(data.index, data['JPM'], label = 'JPM: J P Morgan', linewidth = 2)
plt.plot(data.index, data['BAC'], label = 'BAC: Bank of America', linewidth = 2)
plt.plot(data.index, data['C'], label = 'C: Citi Group', linewidth = 2)
plt.plot(data.index, data['MS'], label = 'MS: Morgan Stanley', linewidth = 2)
plt.plot(data.index, data['WFC'], label = 'WFC: Wells Fargo', linewidth = 2)
plt.plot(data.index, data['SPY'], label = 'SPY: S&P 500 ETF', linewidth = 2)

plt.title('Stock Price Comparison: Major Banks vs S&P 500')
plt.xlabel('Date')
plt.ylabel('Close Price ($)')
plt.grid(True, linestyle = '--', alpha = 0.5)
plt.legend(fontsize = 12, loc = 'upper left')
```

```python
# Feature Engineering
df = pd.DataFrame(index = data.index)

# Create Lag Features - Stock Data
df['BAC(t-1)'] = data['BAC'].shift(1)
df['C(t-1)'] = data['C'].shift(1)
df['MS(t-1)'] = data['MS'].shift(1)
df['WFC(t-1)'] = data['WFC'].shift(1)
df['SPY(t-1)'] = data['SPY'].shift(1)

# Create Lag Features - Stock Data
df['VIX(t-1)'] = data['^VIX'].shift(1)
df['10Y_Yield(t-1)'] = data['^TNX'].shift(1)
df['Gold_Futures(t-1)'] = data['GC=F'].shift(1)
df['US_Dollar_Index(t-1)'] = data['DX-Y.NYB'].shift(1)
df['Crude_Oil_Futures(t-1)'] = data['CL=F'].shift(1)

# Technical Indicators = Moving Average and Rolling Volatility
df['JPM_MA5'] = data['JPM'].rolling(window = 5).mean().shift(1)
df['JPM_MA10'] = data['JPM'].rolling(window = 10).mean().shift(1)
df['JPM_Volatility5'] = data['JPM'].pct_change(5).shift(1)

df['Target'] = data['JPM']

df = df.dropna()
df
```

```python
X = df.drop('Target', axis = 1)
Y = df['Target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, shuffle = False, test_size = 0.1)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(max_depth = 4)
dt_model.fit(X_train,Y_train)

from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators = 100)
rf_model.fit(X_train,Y_train)

from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor(n_neighbors = 5)
knn_model.fit(X_train,Y_train)

from sklearn.svm import SVR
svr_model = SVR()
svr_model.fit(X_train,Y_train)
```

```python
dt_pred = dt_model.predict(X_test)
rf_pred = rf_model.predict(X_test)
knn_pred = knn_model.predict(X_test)
svr_pred = svr_model.predict(X_test)
```

```python
result = pd.DataFrame(Y_test.index)
result['Actual'] = Y_test.values
result['DT Prediction'] = dt_pred
result['RF Prediction'] = rf_pred
result['KNN Prediction'] = knn_pred
result['SVR Prediction'] = svr_pred
result
```

```python
from sklearn.metrics import r2_score, mean_squared_error

def evaluate_model(y_true, y_pred, model_name):
    r2 = r2_score(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    print("Model Name:", model_name)
    print("R2 Value:", r2)
    print("MSE:", mse)
    print("RMSE:", rmse)
    print("\n")

evaluate_model(Y_test, dt_pred, "Decision Tree")
evaluate_model(Y_test, rf_pred, "Random Forest")
evaluate_model(Y_test, knn_pred, "K Nearest Neighbor")
evaluate_model(Y_test, svr_pred, "Support Vector Regressor")
```

```python
plt.figure(figsize = (14,8))
plt.plot(Y_test.index, Y_test, label = 'Actual JPM Stock Price', linewidth = 2, linestyle ='--')
plt.plot(Y_test.index, dt_pred, label = 'Decision Tree Prediction', linewidth = 2, linestyle ='--')
plt.plot(Y_test.index, rf_pred, label = 'Random Forest Prediction', linewidth = 2, linestyle ='--')
plt.plot(Y_test.index, knn_pred, label = 'K Nearest Neighbor Prediction', linewidth = 2, linestyle ='--')
plt.plot(Y_test.index, svr_pred, label = 'Support Vector Regressor Prediction', linewidth = 2, linestyle ='--')

plt.title("Actual vs Predicted JPM Stock Price")
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.grid(alpha = 0.5)
plt.legend()
plt.show()
```
