```python
import yfinance as yf
import pandas as pd
tickers = ['INFY.BO','ONGC.BO','SBIN.BO','TCS.BO','RELIANCE.BO','NTPC.BO','LT.BO','HDFCBANK.BO', 'ICICIBANK.BO', 'ITC.BO']

adj_close_df = pd.DataFrame()

for i in tickers:
    data = yf.download(i, period = '10y', interval = '1d')
    adj_close_df[i] = data ['Adj Close']
```

```python
#Stock Prices

import matplotlib.pyplot as plt

plt.figure(figsize=(20,9))

col = list(adj_close_df.columns)

for i in col:
    plt.plot(adj_close_df.index,adj_close_df[i])
plt.ylabel('Price')
plt.xlabel('Years')
plt.title('Stocks price of past 10 years')
plt.grid()
plt.legend(col)
plt.show()
```

```python
# Risk Return Analysis
col
```

```python
for i in col:
    adj_close_df[i + 'Returns%'] = adj_close_df[i]/adj_close_df[i].shift(1) - 1
```

```python
adj_close_df.dropna(inplace = True)
adj_close_df
```

```python
list(adj_close_df.columns[-10:])
```

```python
import numpy as np

return_columns = list(adj_close_df.columns[-10:])
return1 =[]
risk1 =[]
for i in return_columns:
    temp = adj_close_df[i].mean()
    temp2= ((temp + 1)**250 -1) * 100
    return1.append(temp2)
    
    temp3 = (adj_close_df[i].std() * np.sqrt(250)) * 100
    risk1.append(temp3)
```

```python
return1
```

```python
risk1
```

```python
risk_return_df = pd.DataFrame(index = col)
risk_return_df['Risk (%)'] = risk1
risk_return_df['Return (%)'] = return1
risk_return_df['Risk Adjusted Return'] = (risk_return_df['Risk (%)']/risk_return_df['Return (%)'])
```

```python
risk_return_df
```

```python
# ONGC is the BEST PERFORMED STOCK and HDFC BANK is the WORST PERFORMED STOCK
```

```python
risk_return_list = list(risk_return_df['Risk Adjusted Return'])
risk_return_list
```

```python
sum_risk_list = np.sum(risk_return_list)
sum_risk_list
```

```python
weight_array = risk_return_list/sum_risk_list
weight_array
```

```python
weight_array.sum()
```

```python
risk_return_df['Weightage'] = weight_array
risk_return_df
```

```python
adj_close_df
```

```python
return_df = adj_close_df.iloc[:,10:]
return_df
```

```python
return_df.columns
```

```python
rebased_df = pd.DataFrame()

for j in list(return_df.columns):
    rebased_price_list = [100]
    for i in list(return_df[j]):
        temp = rebased_price_list[-1] * (i+1)
        rebased_price_list.append(temp)
    rebased_df[j] = rebased_price_list
```

```python
rebased_df
```

```python
plt.plot(rebased_df)
plt.legend(rebased_df.columns)
plt.show()
```

```python
rebased_df.dot(list(risk_return_df['Weightage']))
```

```python
final_df = pd.DataFrame()
final_df['Portfolio NAV'] = rebased_df.dot(list(risk_return_df['Weightage']))
```
